```python
import seaborn as sns
import matplotlib.pyplot as plt

```

## Liczba osób z depresją

```python
depression_counts = train_df['Depression'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(depression_counts, labels=['Bez depresji', 'Z depresją'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set1"))
plt.title('Rozkład depresji w zbiorze danych')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(data=train_df, x='Depression', palette='coolwarm')
plt.title('Liczba osób z depresją i bez depresji')
plt.xlabel('Depresja (0 = Bez depresji, 1 = Z depresją)')
plt.ylabel('Liczba osób')
plt.show()
```

## Depresja a płeć

```python

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.countplot(data=train_df, x='Gender', hue='Depression', palette='Blues')


plt.title('Liczba osób z depresją według płci')
plt.xlabel('Płeć')
plt.ylabel('Liczba osób')
plt.legend(title='Depresja', labels=['Brak', 'Depresja'])
plt.show()
```

## Depresja a wiek

```python

train_df['Depression_Label'] = train_df['Depression'].apply(lambda x: 'Bez depresji' if x == 0 else 'Z depresją')




plt.figure(figsize=(10, 6))
sns.histplot(data=train_df, x='Age', hue='Depression_Label', kde=True, palette='Set2', bins=30)
plt.title('Histogram wieku osób z depresją i bez depresji')
plt.xlabel('Wiek')
plt.ylabel('Liczba osób')
plt.legend(title='Depresja', labels=['Z depresją','Bez Depresji'])
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=train_df, x='Depression_Label', y='Age', palette='Set2')
plt.title('Wiek osób z depresją i bez depresji')
plt.xlabel('Depresja')
plt.ylabel('Wiek')
plt.show()

```

## Depresja a życie zawodowe lub studenckie

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Wykres słupkowy dla zmiennej 'Working Professional or Student' z podziałem na 'Depression'
plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Working Professional or Student', hue='Depression', palette='Set3')
plt.title('Rozkład depresji wśród osób pracujących i studentów')
plt.xlabel('Status (Pracujący / Student)')
plt.ylabel('Liczba osób')
plt.legend(title='Depresja', labels=['Bez depresji', 'Z depresją'])
plt.show()

# Definicja zmiennych do analizy
variables = ['Work Pressure', 'Academic Pressure', 'Job Satisfaction', 'Study Satisfaction']
palettes = ['Set2', 'coolwarm', 'viridis', 'cubehelix']

# Tworzenie wykresów pudełkowych (boxplot) dla każdej zmiennej w zmiennych
for var, palette in zip(variables, palettes):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=train_df, x='Depression', y=var, palette=palette)
    plt.title(f'{var} a depresja')
    plt.xlabel('Depresja (0 = Bez depresji, 1 = Z depresją)')
    plt.ylabel(var)
    plt.show()

```

```python

expected_categories = ['Healthy', 'Moderate', 'Unhealthy']
df_filtered = train_df[train_df['Dietary Habits'].isin(expected_categories)]


plt.figure(figsize=(10, 6))
sns.countplot(data=df_filtered, x='Dietary Habits', hue='Depression', palette='Set1')
plt.title('Rozkład nawyków żywieniowych wśród osób z depresją i bez depresji')
plt.xlabel('Nawyki żywieniowe')
plt.ylabel('Liczba osób')
plt.legend(title='Depresja', labels=['Bez depresji', 'Z depresją'])
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=train_df, x='Family History of Mental Illness', hue='Depression', palette='Accent')
plt.title('Rozkład historii chorób psychicznych w rodzinie wśród osób z depresją i bez depresji')
plt.xlabel('Historia chorób psychicznych w rodzinie')
plt.ylabel('Liczba osób')
plt.legend(title='Depresja', labels=['Bez depresji', 'Z depresją'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=train_df, x='Depression', y='Financial Stress', palette='viridis')
plt.title('Rozkład stresu finansowego wśród osób z depresją i bez depresji')
plt.xlabel('Depresja (0 = Bez depresji, 1 = Z depresją)')
plt.ylabel('Financial Stress')
plt.show()
```

## Korelacje

```python
# Wybór tylko kolumn numerycznych
numerical_df = train_df.select_dtypes(include=['float64', 'int64'])

# Obliczanie korelacji między kolumnami numerycznymi
correlation_matrix = numerical_df.corr()

# Tworzenie heatmapy korelacji
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmapa korelacji między zmiennymi numerycznymi')
plt.show()
```
