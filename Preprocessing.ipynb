```python
import matplotlib.pyplot as plt


city_counts = train_df['City'].value_counts().head(10)

plt.figure(figsize=(10, 6))
plt.bar(city_counts.index, city_counts.values, width=0.6)


plt.title("Top 10 Cities by Count", fontsize=16)
plt.xlabel("City", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45, fontsize=12)

plt.tight_layout()


plt.show()
```

```python

city_total_count = train_df['City'].value_counts()


city_depression_count = train_df[train_df['Depression'] == 1]['City'].value_counts()


city_ratio = city_depression_count / city_total_count


train_df['Depression_Ratio_For_City'] = train_df['City'].map(city_ratio)

```

```python
train_df.head(7)
```

```python
# Ensure city_counts contains numeric data
city_counts = train_df['City'].value_counts()

# Filter cities where the count is greater than 500
filtered_cities = city_counts[city_counts > 500].index

# Filter the depression ratios for these cities
filtered_ratios = train_df[train_df['City'].isin(filtered_cities)].groupby('City')['Depression_Ratio_For_City'].mean()

# Sort by depression ratio and select the top 10
top_cities = filtered_ratios.sort_values(ascending=False).head(10)

# Plot the data
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(top_cities.index, top_cities.values, color='skyblue', width=0.6)

# Add labels and title
plt.title("Top 10 Cities by Depression Ratio (City Count > 500)", fontsize=16)
plt.xlabel("City", fontsize=14)
plt.ylabel("Depression Ratio", fontsize=14)
plt.xticks(rotation=45, fontsize=12)

plt.tight_layout()

# Show the plot
plt.show()
```
